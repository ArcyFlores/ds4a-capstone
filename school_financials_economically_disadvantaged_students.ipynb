```python
import pandas as pd
import numpy as np
df_dropout=pd.read_csv('data/acgr19.csv')
```

```python
df_dropout.head()
```

```python
df_dropout = df_dropout.dropna(subset=['DistrictCode'])
df_dropout['DistrictCode'] = df_dropout['DistrictCode'].astype(int).astype(str)
```

```python
df_UserGL=pd.read_csv('data/UserGL.csv')
```

```python
df_UserGL.head()
 


```

```python
# Group df_UserGL by Dcode, create a dictionary
fund_by_district=df_UserGL.groupby('Dcode')['Value'].sum()
```

```python
average_dropout_rate_by_district=df_dropout.groupby('DistrictCode')['Dropout (Rate)']
```

```python
#average_dropout_rate_by_district=df_dropout.groupby('DistrictCode')['Dropout (Rate)'].mean()
```

```python
df_dropout['Dropout (Rate)'] = pd.to_numeric(df_dropout['Dropout (Rate)'], errors='coerce')
```

```python
df_dropout_clean = df_dropout.dropna(subset=['Dropout (Rate)'])
average_dropout_rate_by_district = df_dropout_clean.groupby('DistrictCode')['Dropout (Rate)'].mean()
```

```python
avg_dropout_rate_by_district = pd.DataFrame({'Dcode': average_dropout_rate_by_district.index, 'avg_dropout_rate': average_dropout_rate_by_district.values})

total_fund_by_district=pd.DataFrame({'Dcode': fund_by_district.index, 'total_fund':fund_by_district.values})
```

```python
total_fund_by_district['Dcode']=total_fund_by_district['Dcode'].astype(int).astype(str)
```

```python
merged_df = pd.merge(avg_dropout_rate_by_district, total_fund_by_district, on='Dcode')
```

```python
merged_df.head(20)
```

```python
df_expense=pd.read_csv('data/currentexpense1819_by_district.csv')
df_expense.head()
```

```python
# Check for duplicate Dcode in df_expense
if df_expense['Dcode'].duplicated().any():
    print("There are duplicate Dcode in df_expense")
else:
    print("There are no duplicate Dcode in df_expense")

df_expense['Dcode']=df_expense['Dcode'].astype(int).astype(str)
```

```python
dropout_fund_cost = pd.merge(merged_df, df_expense, on='Dcode')

dropout_fund_cost.head()
```

```python
dropout_fund_cost.to_csv('dropout_fund_cost.csv')
```

```python
df_detail_fund=pd.read_csv('data/detail_fund_cost.csv')
df_detail_fund.head()
```

```python
d1 = df_detail_fund['District'].tolist()
d2 = dropout_fund_cost['District'].tolist()

import numpy as np
d1 = [_.replace('-', ' ').replace('/', ' ') for _ in d1]
d2 = [_.replace('-', ' ').replace('/', ' ') for _ in d2]


pairs = []
for i in range(len(d2)):
    find_js = [0 for _ in range(len(d1))]
    for j in range(len(d1)):
        # if d2[i].lower() in d1[j].lower() or d1[j].lower() in d2[i].lower():
        if d2[i].lower().split()[0] != d1[j].lower().split()[0]:
            continue
        for k in range(min(len(d1[j].split()), len(d2[i].split()))):
            if d2[i].lower().split()[k] ==  d1[j].lower().split()[k] or (d2[i].lower().split()[k] in ['uni', 'unif', 'unified'] and d1[j].lower().split()[k] in ['uni', 'unif', 'unified']) :
                find_js[j] += 1
    final_j = np.argmax(find_js)
    if find_js[final_j] <= 0:
        print('cannot find 0', d2[i])
        for j in range(len(d1)):
            for k in range(min(len(d1[j].split()), len(d2[i].split()))):
                if d2[i].lower().split()[k] ==  d1[j].lower().split()[k] or (d2[i].lower().split()[k] in ['uni', 'unif', 'unified'] and d1[j].lower().split()[k] in ['uni', 'unif', 'unified']) :
                    find_js[j] += 1     
        final_j = np.argmax(find_js)
        if find_js[final_j] <= 0:
            print('cannot find', d2[i])
   
    pairs.append([d2[i], d1[final_j]])

new_name = []
for p in pairs:
    # print(p)
    new_name.append(p[1])

dropout_fund_cost['District'] = new_name
dropout_fund_cost
```

```python
dropout_fund_cost_detail = pd.merge(dropout_fund_cost, df_detail_fund, on='District')
```
