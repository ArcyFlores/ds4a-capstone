```python
import pandas as pd
import numpy as np
```

### Split the large file into smaller ones
- The number of rows is too large to be loaded into csv files
- There will be encoding issues if directly read from txt files

```python
with open("tamo2021.txt",'r') as file:
    lines = file.readlines()

with open("tamo2021_A.txt",'w') as file:
    for line in lines[:int(len(lines)/2)]:
        file.write(line)

with open("tamo2021_B.txt",'w') as file:
    for line in lines[int(len(lines)/2):]:
        file.write(line)
```

### Manually load txt data into excel files and save as csv files

### Load the data
- The staff data at different granularity levels: school, school district, county, state

```python
df_staff_a = pd.read_csv('tamo2021_A.csv', low_memory=False)
df_staff_a.head()
```

```python
df_staff_b = pd.read_csv('tamo2021_B.csv', low_memory=False)
df_staff_b.head()
```

### Withdraw the teaching assignment outcome with filters
- Aggregate Level = S. School level data
- School Grade Span = GR912. Grade 9â€“12 public high schools
- Teacher Experience Level = All. All the teachers
- Teacher Credential Level = All. All the teachers
- Subject Area = MATH, SCIE, ENLA, HISS. Most important school subjects I believe. MATH = Mathematics, SCIE = Science, ENLA = English Language Arts, HISS = History/Social Science

```python
df_staff_a_filtered = df_staff_a[(df_staff_a['Aggregate Level'] == 'S') & (df_staff_a['School Grade Span'] == 'GR912')
                                & (df_staff_a['Teacher Experience Level'] == 'ALL') & (df_staff_a['Teacher Credential Level'] == 'ALL')]
df_staff_b_filtered = df_staff_b[(df_staff_b['Aggregate Level'] == 'S') & (df_staff_b['School Grade Span'] == 'GR912')
                                & (df_staff_b['Teacher Experience Level'] == 'ALL') & (df_staff_b['Teacher Credential Level'] == 'ALL')]
```

```python
df_staff_a_filtered
```

```python
df_staff_b_filtered.head()
```

```python
df_staff_a_filtered.to_csv('tamo2021_A_filtered.csv', encoding='utf-8', index=False)
df_staff_b_filtered.to_csv('tamo2021_B_filtered.csv', encoding='utf-8', index=False)
```

```python
# Combine the two sets of data
combined = pd.concat([df_staff_a_filtered, df_staff_b_filtered])
combined.to_csv('tamo2021_school_level.csv', encoding='utf-8', index=False)
```

### The important features include
- Total FTE (count). FTE: Full-Time Equivalent
- Clear FTE (count). Where the assignment monitoring outcome is clear
- Out-of-Field FTE (count). Where the assignment monitoring outcome is out-of-field
- Intern FTE (count). Where the assignment monitoring outcome is intern
- Ineffective FTE (count). Where the assignment monitoring outcome is ineffective
- Incomplete FTE (count). Where the assignment monitoring outcome is incomplete
- Unknown FTE (count). Where the assignment monitoring outcome is unknown
- N/A FTE (count). Where the assignment monitoring outcome is N/A
#### Notes: there are also percentage data like Clear FTE (percent), Clear FTE (count) divided by the Total FTE (count)
